#### 2. Cleaning the data-  
There are some indicators which are not updated recently. The percentage of the youth rate, the male rate and the female rate 
of the age 15-24 is available till the year 2011. The male mean and the female mean of the age 15-24 is available till the year 2015.The data on expenditure spent by Government on per student is available till the year 2018.
I'm choosing the year 2011 as a common year in all the indicators and changing their column names then merging the multiple dataframe in a single dataframe.

```python
#selection the year 2011 and changing the column name
popn=population[['country','2011']]
popn=popn.rename(columns={'2011':'Population'})

inc=income[['country','2011']]
inc=inc.rename(columns={'2011':'GDP'})

rate=youth_rate[['country','2011']]
rate=rate.rename(columns={'2011':'Youth_Rate'})

male=male_rate[['country','2011']]
male=male.rename(columns={'2011':'Male_Rate'})

female=female_rate[['country','2011']]
female=female.rename(columns={'2011':'Female_Rate'})

m_mean=male_mean[['country','2011']]
m_mean=m_mean.rename(columns={'2011':'Male_Mean'})

f_mean=female_mean[['country','2011']]
f_mean=f_mean.rename(columns={'2011':'Female_Mean'})

expend=expenditure[['country','2011']]
expend=expend.rename(columns={'2011':'Expenditure'})
```

```python
#merging the data
data=(popn.merge(inc,how='outer').merge(rate,how='outer').merge(male,how='outer').merge(female,how='outer').merge(m_mean,how='outer').merge(f_mean,how='outer').merge(expend,how='outer'))
```

> As it is difficult to find the missing values and analysis the missing values of every indicator in the year 2011, I'm filtering the data

```python
#Filtering the rows which has missing values
data=data.dropna()
```
